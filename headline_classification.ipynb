```python
headlines_df['classification'] = ''
classification_lst = []

def headline_classifier(string, keywords_tuple, name_class_tuple):
    string = string.lower()
    if len(keywords_tuple) == len(name_class_tuple) and len(keywords_tuple) > 0:
        no_of_classes = len(keywords_tuple)
        counts = []
        count = 0
        while no_of_classes > 0:
            no_of_classes -= 1
            for keyword in keywords_tuple[no_of_classes]:
                if keyword in string:
                    count += 1
            counts.insert(0, count)
            count = 0
    else:
        return "List of keywords do not match list of classes or the list of keywords is empty"
        
    if max(counts) > 0:
        max_list = max(counts)
        values = np.array(counts)
        ii = np.where(values==max_list)[0]
        results_temp = [name_class_tuple[index] for index in ii]
        sep = ","
        results = sep.join(results_temp)
        return results
    else:
        return 'Other'
        
def headline_classification(data, keywords_tuple, name_class_tuple):
    n_row, n_col = data.shape
    data_lst = list(data['headline'].values)

    #Headline classification
    classification_lst = []
    for i in range(n_row):
        data_i = str(data_lst[i]) #Include this becasue of potential nan values
        classification_lst.append(headline_classifier(data_i, keywords_tuple, name_class_tuple))
    return classification_lst
```

```python
lst_1 = headline_classification(headlines_df, (british_keywords, climate_keywords), ('Brexit', 'Climate'))
headlines_df = headlines_df.assign(classification=lst_1)
```

# Sentimental Analysis

```python
def sentiments(data):
    n_row, n_col = data.shape
    data_lst = list(data['headline'].values)

    #Empty list to add the polarity score 
    polarity_lst = []
    subjectivity_lst = []

    #Headline sentiment
    for i in range(n_row):
        headline_i = str(data_lst[i]) #Include this becasue of potential nan values
        blob_headline_i = TextBlob(headline_i) #transforming string into textblob
        polarity_lst.append(blob_headline_i.sentiment.polarity)
        subjectivity_lst.append(blob_headline_i.sentiment.subjectivity)
    
    return (polarity_lst, subjectivity_lst)
```

```python
brexit_sentiments = sentiments(brexit_dataset)
brexit_dataset = brexit_dataset.assign(polarity=brexit_sentiments[0], subjectivity=brexit_sentiments[1])
```

```python
brexit_dataset
```

```python
climate_dataset = headlines_df[headlines_df['classification'].str.contains('Climate')]
other_dataset = headlines_df[headlines_df['classification'].str.contains('Other')]
```

```python
climate_sentiments = sentiments(climate_dataset)
climate_dataset = climate_dataset.assign(polarity=climate_sentiments[0], subjectivity=climate_sentiments[1])
```

```python
other_sentiments = sentiments(other_dataset)
other_dataset = other_dataset.assign(polarity=other_sentiments[0], subjectivity=other_sentiments[1])
```

```python
results_df = pd.DataFrame(index = ['polarity', 'subjectivity'], columns = ['Brexit', 'Climate', 'Brexit, Climate', 'Brexit Immigration', 'Brexit Other', 'Other'])
only_brexit = brexit_dataset[brexit_dataset['classification'] == 'Brexit']
only_climate = climate_dataset[climate_dataset['classification'] == 'Climate']
both = brexit_dataset[brexit_dataset['classification'] == 'Brexit, Climate']
brexit_imm = only_brexit[only_brexit['brexit'] == "Immigration"]
brexit_other = only_brexit[only_brexit['brexit'] == "Other"]

results_df['Brexit'] = only_brexit.mean(axis = 0, numeric_only = True)
results_df['Climate'] = only_climate.mean(axis = 0, numeric_only = True)
results_df['Brexit Immigration'] = brexit_imm.mean(axis = 0, numeric_only = True)
results_df['Brexit Other'] = brexit_other.mean(axis = 0, numeric_only = True)
results_df['Brexit, Climate'] = both.mean(axis = 0, numeric_only = True)
results_df['Other'] = other_dataset.mean(axis = 0, numeric_only = True)
```
